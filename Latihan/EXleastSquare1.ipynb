```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 9, 11]

x = np.array(x)
y = np.array(y)

A = np.vstack([x, np.ones(len(x))]).T
y = y[:, np.newaxis]

alpha = np.linalg.inv(np.dot(A.T, A)).T
alpha = np.dot(alpha, A.T)
alpha = np.dot(alpha, y)

print(f"Alpha (m dan c): {alpha[0]}")
print(alpha[1])

plt.plot(x, y, "ro")
plt.plot(x, alpha[0]*x+alpha[1], 'b')
plt.show()
```
