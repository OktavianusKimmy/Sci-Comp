GAUSS SEIDEL

```python
import numpy as np
```

```python
x_arr = [
    [
        [8, 4, 2],
        [1, 6, 3],
        [5, 6, 12]
    ],
    [
        [12, -7, -1],
        [1, -7, 2],
        [8, -1, 12]
    ],
    [
        [4, 2, 7],
        [3, -7, 5],
        [2, -1, 3]
    ],
    [   
        [14, 2, 1],
        [1, 8, 4],
        [9, 8, 18]
    ],
    [   
        [1, 3, 6],
        [9, 1, 12],
        [15, 18, 1]
    ]
]
y_arr = [
    [10, 12, 15],
    [21, 12, 4],
    [8, 4, 5],
    [7, 2, 6],
    [4, 7, 6]
]
```

```python
def check_diagonal_dominant(x):
    x = np.array(x)
    diag = np.diag(np.abs(x))
    non_diag = np.sum(np.abs(x), axis=1)
    if np.all(diag > non_diag - diag):
        return True
    else:
        return False
    
def gauss_seidel(x, y, max_iter = 20, tol = 0.055):
    x = np.array(x)
    y = np.array(y)
    diag = np.diag(x)
    x = -x
    np.fill_diagonal(x, 0)
    old_res = np.zeros(len(x[0]))

    for i in range(max_iter):
        new_res = np.array(old_res)
        for j, row in enumerate(x):
            new_res[j] = (y[j] + np.dot(row, new_res))
        print(f"Iteration #{i+1} : {new_res}")

        distance = np.sqrt(np.dot(old_res - new_res, old_res - new_res))
        print(f"Difference = {distance}")
        
        if distance < tol:
            return True
        
        old_res = new_res
    return False

def run(x, y):
    for i, a in enumerate(x):
        print(f"Matrix #{i+1}")
        if check_diagonal_dominant(a):
            print("Diagonal dominant")
            if gauss_seidel(a, y[i]):
                print("Convergent reached")
            else:
                print("Max iter reached")
        else:
            print("Not diagonal dominant")
```

```python
run(x_arr, y_arr)
```
