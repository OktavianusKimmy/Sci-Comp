```python
import numpy as np
```

```python
x_arr = [
    [
        [5, 1, 2],
        [2, 6, 1],
        [1, 1, 7]
    ],
    [
        [10, 2, -1],
        [1, 8, 2],
        [2, -1, 9]
    ],
    [
        [6, -1, 1],
        [3, 7, 2],
        [2, -2, 8]
    ],
    [
        [4, 2, 1],
        [1, 5, 2],
        [3, 1, 6]
    ],
    [
        [8, -1, 2],
        [1, 9, 1],
        [2, 3, 10]
    ]
]
y_arr = [
    [12, 14, 17],
    [13, 18, 16],
    [11, 19, 14],
    [10, 13, 15],
    [17, 14, 19]
]
```

```python
def check_diagonal_dominant(x):
    x = np.array(x)
    diag = np.diag(np.abs(x))
    non_diag = np.sum(np.abs(x), axis=1)
    if np.all(diag > non_diag - diag):
        return True
    else:
        return False
    
def gauss_seidel(x, y, max_iter = 45, tol = 0.001):
    x = np.array(x)
    y = np.array(y)
    diag = np.diag(x)
    x = -x
    np.fill_diagonal(x, 0)
    old_res = np.zeros(len(x[0]))

    for i in range(max_iter):
        new_res = np.array(old_res)
        for j, row in enumerate(x):
            new_res[j] = (y[j] + np.dot(row, new_res)) / diag[j]
        print(f"Iteration {i+1} : {new_res}")

        distance = np.sqrt(np.dot(old_res - new_res, old_res - new_res))
        if distance < tol:
            print("Convergence")
            print(f"Result: {new_res}")
            return True
        
        old_res = new_res
    return False

def run(x, y):
    for i, a in enumerate(x):
        print(f"Matrix #{i+1}:")
        if check_diagonal_dominant(a):
            print("Diagonally Dominant")
            if not gauss_seidel(a, y[i]):
                print("Max iter reached!")
                
        else:
            print("Not diagonally dominant")
```

```python
run(x_arr, y_arr)
```
