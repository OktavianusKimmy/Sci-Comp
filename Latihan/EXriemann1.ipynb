RIEMANN

```python
import numpy as np
```

```python
def f(x):
    return x**2 + 2*x + 1
```

```python
a = 1
b = 5
points = 1000
width = (b - a) / (points - 1)
x = np.linspace(a, b, points)
print(x)
```

```python
rightRiemann = width * np.sum(f(x[1:]))
print(f"Right: {rightRiemann}")

mid = (x[1:] + x[:-1]) / 2
middleRiemann = width * np.sum(f(mid))
print(f"Middle: {middleRiemann}")
```
